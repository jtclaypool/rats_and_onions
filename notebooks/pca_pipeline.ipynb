```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from massholes.loader import loader
```

```python
path = "../data/rats-and-onions.xls"
```

```python
X, y = loader(path)
```

```python
pipeline = Pipeline([
    #('scaler', ParetoScaler),
    ('pca', PCA(n_components=2))
])

X_fit = pipeline.fit_transform(X.drop("index", axis = 1))
```

```python
% matplotlib inline
import matplotlib.pyplot as plt

symbols = ["o", ">", "x", "+"]
groups = pd.DataFrame(data = X_fit, columns = ["pca1", "pca2"]).groupby(y) 
fig, ax = plt.subplots()

for i, group in enumerate(groups):
    group[1].plot(x = "pca1", y = "pca2", kind = "scatter", marker = symbols[i], ax = ax)
```
